# Multiple seasons

```python


leagues=['E0','E2','E3']
frames = []
#loop through leagues read multiple csv and append into a list
for league in leagues:
    for season in range(15,24):
        df= pd.read_csv(root+str(season)+str(season+1) + '/' +  league +'.csv',encoding = 'unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
```

# 3 Organize all the data in dictionary

```python
dist_countries = {'Spanish La Liga' : 'sp1',
                  'Spanish Segunda Division':'sp2',
                  'German Bundesliga':'d1',
                  'English Premier League':'e1',
                  'English Championship':'e2'}
```

```python
dist_countries['Spanish La Liga']
```

```python

dist_hist={};

#loop through leagues read multiple csv and append into a list
for league in dist_countries:
    frames = []
    for season in range(15,24):
        df= pd.read_csv(root+str(season)+str(season+1) + '/' +  dist_countries[league] +'.csv',encoding = 'unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dist_hist[league] = df_concat
    
```
