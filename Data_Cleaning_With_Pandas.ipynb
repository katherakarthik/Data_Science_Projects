# Dealing with missing data

```python
# remove a column or row with dropna or drop or isnull
#drop column
# df_netflix.drop('director',axis=1)
# no_director=df_netflix[df_netflix['director'].isnull()].index
# df_netflix.drop(no_director,axis=0)
# ~ +.isnull()
# df_netflix[~(df_netflix['director'].isnull())]
#dropna()
df_netflix.dropna(subset=['director'])
```

```python
# Use fillna to replace NAN by mean,median numeric data or mode
mode=''.join(df_netflix['rating'].mode())
df_netflix['rating'].fillna(mode,inplace=True)
# df_netflix['rating'].fillna()
```

```python
df_netflix.isnull().mean()
```

```python
# Replace it by an arbitrary number with .fillna()
# df_netflix.fillna({'duration': 0}, inplace=True)
#backward,forward
df_netflix['director'].fillna(method='bfill')
```

```python
#Extracting data from duration column with split() and extract method
# 1. Work on an explicit copy
df_movie = df_netflix[df_netflix['type'] == 'Movie'].copy()

# 2. Extract the numeric part of the duration string
df_movie['duration_min'] = (df_movie['duration']
                              .str.extract(r'(\d+)')      # first run of digits
                              .astype(float)              # temporarily float to handle NaNs
                              .fillna(0)                  # replace any remaining NaNs
                              .astype(int))               # final integer column
```
